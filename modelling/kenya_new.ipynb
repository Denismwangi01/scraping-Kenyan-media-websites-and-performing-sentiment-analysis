```python
data.dropna(inplace=True)
```

```python
import pandas as pd
for col in ['publication_date', 'created_at', 'last_updated']:
    data[col] = pd.to_datetime(data[col], errors='coerce')
data[['publication_date', 'created_at', 'last_updated']] = data[['publication_date', 'created_at', 'last_updated']].fillna(method='ffill')
```

```python
import re

def clean_text(text):
    if pd.isna(text):
        return ""
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = text.lower().strip()
    return text

data['cleaned_content'] = data['content'].apply(clean_text)
```

```python
data.head(15)
```

```python
from textblob import TextBlob

# Function to get polarity score
def get_polarity(text):
    return TextBlob(text).sentiment.polarity

data['polarity'] = data['cleaned_content'].apply(get_polarity)
```

```python
def get_sentiment_label(score):
    if score > 0.1:
        return 'positive'
    elif score < -0.1:
        return 'negative'
    else:
        return 'neutral'

data['sentiment'] = data['polarity'].apply(get_sentiment_label)
```

```python
import pandas as pd
from collections import Counter
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import string

stop_words = set(stopwords.words('english'))
punctuation = set(string.punctuation)

# Optional: Map polarity score to sentiment category
def get_sentiment(polarity):
    if polarity > 0.1:
        return 'positive'
    elif polarity < -0.1:
        return 'negative'
    else:
        return 'neutral'

data['sentiment'] = data['polarity'].apply(get_sentiment)


```

```python
# Tokenize cleaned_content into words
all_rows = []
for _, row in data.iterrows():
    sentiment = row['sentiment']
    words = row['cleaned_content'].lower().split()
    words = [w.strip(string.punctuation) for w in words if w not in stop_words and w.isalpha()]

    for word in words:
        all_rows.append((word, sentiment))

# Create a DataFrame of word-sentiment pairs
word_df = pd.DataFrame(all_rows, columns=['word', 'sentiment'])

# Count frequency of each word per sentiment
word_counts = word_df.groupby(['word', 'sentiment']).size().reset_index(name='count')

# Filter out rare words
word_counts = word_counts[word_counts['count'] >= 3]

# Save
word_counts.to_csv('sentiment_counts.csv', index=False)
```
